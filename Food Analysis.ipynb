```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
df=pd.read_csv('event.csv')
data1=data['NAME'].astype(str)
data2=data['RATING'].astype(str)
plt.figure(figsize=(20,9))
sns.lineplot(x=data1,y=data2,data=df)
plt.show()
```
